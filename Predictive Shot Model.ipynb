```python
import requests

# find game ids on NHL.com, all games from start of 2023 season to all games played yesterday
start_game_number = 2023020001
end_game_number = 2023020836

# Container to store data
all_data = []

# Loop through the range and make requests
for game_number in range(start_game_number, end_game_number + 1):
    url = f"https://api-web.nhle.com/v1/gamecenter/{game_number}/boxscore"
    
    # Make the request
    response = requests.get(url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        all_data.append(response.text)
    else:
        print(f"Error for game {game_number}: {response.status_code}")
```

```python
# Select columns from teams_df that I want to join to the player dataframe
selected_teams_df = teams_df[['team_game_id', 'teams_id', 'game_date']]

# Merge the player_df and teams_df info
merged_df = pd.merge(player_df, selected_teams_df, left_on= ['game_id', 'team_id'], right_on = ['team_game_id', 'teams_id'], how='left')

merged_df
```

```python
# Make sure 'game_date' is in datetime format
merged_df['game_date'] = pd.to_datetime(merged_df['game_date'])

# Sort the DataFrame by 'game_date' to properly get cumalative and rolling averages for each player
merged_df = merged_df.sort_values(by='game_date')

# Calculate the rolling average for the last 5 games for each player
merged_df['rolling_avg_shots'] = (merged_df.groupby('player_id')['shots']
                                              .transform(lambda x: x.shift(1).rolling(window=5, min_periods=1).mean()))
# Calculate the cumlative average for each player
merged_df['cumulative_avg_shots'] = (merged_df.groupby('player_id')['shots']
                                              .transform(lambda x: x.expanding().mean()))
```

```python
# Function to convert 'MM:SS' to total seconds as an ml algortihm will need an integer to correctly interpret
def mmss_to_seconds(mmss):
    minutes, seconds = map(int, mmss.split(':'))
    return minutes * 60 + seconds

# Convert 'toi' to total seconds and create 'toi_seconds' column
merged_df['toi_seconds'] = merged_df['toi'].apply(mmss_to_seconds)

# Convert 'pp_toi' to total seconds and create 'pp_toi_seconds' column
merged_df['pp_toi_seconds'] = merged_df['pp_toi'].apply(mmss_to_seconds)

# Convert 'sh_toi' to total seconds and create 'sh_toi_seconds' column
merged_df['sh_toi_seconds'] = merged_df['sh_toi'].apply(mmss_to_seconds)

# Drop the original 'toi', 'pp_toi', and 'sh_toi' columns
merged_df.drop(columns=['toi', 'pp_toi', 'sh_toi'], inplace=True)
```

```python
# Make sure 'game_date' is in datetime format
merged_df['game_date'] = pd.to_datetime(merged_df['game_date'])

# Sort the DataFrame by 'game_date' to properly calculate rolling and cumalative averages for each game
merged_df = merged_df.sort_values(by='game_date')

# Calculate the rolling average for the last 5 games
merged_df['rolling_avg_toi_seconds'] = (merged_df.groupby('player_id')['toi_seconds']
                                              .transform(lambda x: x.shift(1).rolling(window=5, min_periods=1).mean()))

# Calculate the cumulative average
merged_df['cumulative_avg_toi_seconds'] = (merged_df.groupby('player_id')['toi_seconds']
                                              .transform(lambda x: x.expanding().mean()))

# Calculate the rolling average for the last 5 games
merged_df['rolling_avg_pp_toi_seconds'] = (merged_df.groupby('player_id')['pp_toi_seconds']
                                              .transform(lambda x: x.shift(1).rolling(window=5, min_periods=1).mean()))

# Calculate the cumulative average
merged_df['cumulative_avg_pp_toi_seconds'] = (merged_df.groupby('player_id')['pp_toi_seconds']
                                              .transform(lambda x: x.expanding().mean()))

# Calculate the rolling average for the last 5 games
merged_df['rolling_avg_sh_toi_seconds'] = (merged_df.groupby('player_id')['sh_toi_seconds']
                                              .transform(lambda x: x.shift(1).rolling(window=5, min_periods=1).mean()))

# Calculate the cumulative average
merged_df['cumulative_avg_sh_toi_seconds'] = (merged_df.groupby('player_id')['sh_toi_seconds']
                                              .transform(lambda x: x.expanding().mean()))
```

```python
#insert x and y coordinate shot data for each player and team. this is created from the "Shot Coordinates Final" script and retrieved from NHL API
#We are creating a custom metric to measure the Jaccard Similarity between a players "shot map" and a teams "shot map"
#Jaccard similarity measures the similarity between two lists - the idea is if a player often shoots from an area of the ice where a team often
#gives up a lot of shots, then that player might be more likely to have a higher shot total. The inverse is also true.
#Tested this in the "Shot Coordinates Final" script. The correlation between Jaccard Similarity of a player's shot coordinates and team's shot coordinates
#and their actual shot output for that game was around 0.26 so there is some signal to this approach
#for a high level of summary of Jaccard Similarity check out: https://medium.com/@mayurdhvajsinhjadeja/jaccard-similarity-34e2c15fb524

#read in shot on goal event data
sog_events = pd.read_csv('sog_event_data.csv')

# Convert 'game_date' column to datetime type in both DataFrames
merged_df['game_date'] = pd.to_datetime(merged_df['game_date'])
sog_events['game_date'] = pd.to_datetime(sog_events['game_date'])


#this is the formula to calculate Jaccard similarity between two lists
def jaccard(list1, list2):
    if not list1 or not list2:
        return 0  # Return 0 if either list is empty
    
    intersection = len(set(list1).intersection(list2))
    union = len(set(list1).union(list2))
    
    if union == 0:
        return 0
    else:
        similarity = float(intersection) / union
        distance = 1 - similarity
        return similarity, distance

#convert separate x and y coordinate columns into a single column to compile shot attempt data in one list for players
merged_df['player_shots'] = merged_df.apply(
    lambda row: sog_events.loc[
        (sog_events['player_id'] == row['player_id']) &
        (sog_events['game_date'] < row['game_date'])
    ].assign(xCoord=lambda x: x['xCoord'].abs())[['xCoord', 'yCoord']].astype(str).apply(lambda x: x['xCoord'] + x['yCoord'], axis=1).tolist(),
    axis=1
)

#convert separate x and y coordinate columns into a single column to compile shot attempt data in one list for teams
merged_df['opponent_shots'] = merged_df.apply(
    lambda row: sog_events.loc[
        (sog_events['team_shot_on'] == row['opponent']) &
        (sog_events['game_date'] < row['game_date'])
    ].assign(xCoord=lambda x: x['xCoord'].abs())[['xCoord', 'yCoord']].astype(str).apply(lambda x: x['xCoord'] + x['yCoord'], axis=1).tolist(),
    axis=1
)

#calcualte the Jaccard similarity between a player's shot map and the opponent's shot map
merged_df['jaccard_similarity'] = merged_df.apply(
    lambda row: jaccard(row['player_shots'], row['opponent_shots']),
    axis=1
)


#show dataframe to make sure our new columns look appropriate
merged_df
```

```python
# Our Jaccard formula calculates the Jaccard distance and similarity within a tuple, we want only the similarity
#i.e. the first numerical value in the tuple so this code selects the first element in each tuple
merged_df['jaccard_similarity'] = merged_df['jaccard_similarity'].apply(lambda x: x[0] if x else 0)
```

```python
# train a neural net on the data
# best results so far - have played around with many combinations of hyperparamter tuning by performing grid searches
# The Hyperparameters below have consistently performed the best
# May still play around with number of neurons, number of layers, and activation functions
# Can also train multiple neural nets and create an ensemble approach to improve performance, will explore

import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error


# Define a function to create and train the neural network
def create_and_train_model(X_train, y_train, X_test, y_test, epochs=50, batch_size=32, learning_rate=0.001):
    model = keras.models.Sequential([
        keras.layers.Dense(100, activation="sigmoid", input_shape=X_train.shape[1:]),
        keras.layers.Dropout(0.1),  # Adding dropout layer to reduce overfitting
        keras.layers.Dense(50, activation="tanh"),  # Example of using tanh activation
        keras.layers.Dropout(0.1),  # Adding dropout layer to reduce overfitting
        keras.layers.Dense(50, activation="tanh"),  # Example of using tanh activation
        keras.layers.Dense(1)  # Output layer with one neuron for regression
    ])

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    
    model.compile(optimizer=optimizer, loss="mean_absolute_error")

    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.1, verbose=0)

    mae_test = model.evaluate(X_test, y_test)

    return model, mae_test, history

# Hyperparameter Tuning, may test additional values here as well
learning_rates = [0.001]
batch_sizes = [32]
epochs_list = [100]

best_model = None
best_mae = float('inf')

for lr in learning_rates:
    for bs in batch_sizes:
        for epochs_value in epochs_list:
            model, mae, _ = create_and_train_model(X_train, y_train, X_test, y_test, 
                                                   epochs=epochs_value, batch_size=bs, learning_rate=lr)
            
            print(f"Learning Rate: {lr}, Batch Size: {bs}, Epochs: {epochs_value}, MAE: {mae}")
            
            if mae < best_mae:
                best_model = model
                best_mae = mae


#predict on the test set
y_pred = best_model.predict(X_test)

print("Best Mean Absolute Error:", best_mae)
print("Predicted Output for X_new:", y_pred)
```

```python
#get Jaccard data for players for today's games - for an explanation of this process look above to the prior code cell containing info about
#Jaccard similarity

sog_events = pd.read_csv('sog_event_data.csv')

# Convert 'game_date' column to datetime type in both DataFrames
predictions['game_date'] = pd.to_datetime(predictions['game_date'])
sog_events['game_date'] = pd.to_datetime(sog_events['game_date'])


def jaccard(list1, list2):
    if not list1 or not list2:
        return 0  # Return 0 if either list is empty
    
    intersection = len(set(list1).intersection(list2))
    union = len(set(list1).union(list2))
    
    if union == 0:
        return 0
    else:
        similarity = float(intersection) / union
        distance = 1 - similarity
        return similarity, distance


predictions['player_shots'] = predictions.apply(
    lambda row: sog_events.loc[
        (sog_events['player_id'] == row['player_id']) &
        (sog_events['game_date'] < row['game_date'])
    ].assign(xCoord=lambda x: x['xCoord'].abs())[['xCoord', 'yCoord']].astype(str).apply(lambda x: x['xCoord'] + x['yCoord'], axis=1).tolist(),
    axis=1
)


predictions['opponent_shots'] = predictions.apply(
    lambda row: sog_events.loc[
        (sog_events['team_shot_on'] == row['opponent']) &
        (sog_events['game_date'] < row['game_date'])
    ].assign(xCoord=lambda x: x['xCoord'].abs())[['xCoord', 'yCoord']].astype(str).apply(lambda x: x['xCoord'] + x['yCoord'], axis=1).tolist(),
    axis=1
)


predictions['jaccard_similarity'] = predictions.apply(
    lambda row: jaccard(row['player_shots'], row['opponent_shots']),
    axis=1
)

predictions
```

```python
# Manipulate rows based on the values in the tuple
predictions['jaccard_similarity'] = predictions['jaccard_similarity'].apply(lambda x: x[0] if x else 0)

predictions
```
